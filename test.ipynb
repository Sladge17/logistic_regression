```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
column = ['name', 'numb1', 'numb2']
string1 = ['Ivan', 12, 4567]
string2 = ['Petr', 18, 1245]

df = pd.DataFrame([string1, string2], columns=column)
df.set_index('name', inplace=True)
df
```

```python
list(df.loc['Ivan'])
```

```python
plt.figure(figsize=(15, 8))
plt.bar(range(len(list(df.loc['Ivan']))), list(df.loc['Ivan']))
plt.show()
```
