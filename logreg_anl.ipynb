```python
Ravenclaw = df_train[df_train['Hogwarts House'] == 'Ravenclaw'].drop(['Index', 'Hogwarts House', 'First Name', 'Last Name', 'Birthday', 'Best Hand'], axis=1)
Slytherin = df_train[df_train['Hogwarts House'] == 'Slytherin'].drop(['Index', 'Hogwarts House', 'First Name', 'Last Name', 'Birthday', 'Best Hand'], axis=1)
Gryffindor = df_train[df_train['Hogwarts House'] == 'Gryffindor'].drop(['Index', 'Hogwarts House', 'First Name', 'Last Name', 'Birthday', 'Best Hand'], axis=1)
Hufflepuff = df_train[df_train['Hogwarts House'] == 'Hufflepuff'].drop(['Index', 'Hogwarts House', 'First Name', 'Last Name', 'Birthday', 'Best Hand'], axis=1)
```

```python
Ravenclaw_hist = Ravenclaw.describe().T['count'].tolist()
Slytherin_hist = Slytherin.describe().T['count'].tolist()
Gryffindor_hist = Gryffindor.describe().T['count'].tolist()
Hufflepuff_hist = Hufflepuff.describe().T['count'].tolist()
```

```python
plt.figure(figsize=(15, 8))
x = np.arange(13)
plt.bar(range(len(Ravenclaw_hist)), Ravenclaw_hist, width=1, align='edge', color='blue', alpha=0.5)
plt.bar(range(len(Slytherin_hist)), Slytherin_hist, width=1, align='edge', color='green', alpha=0.5)
plt.bar(range(len(Gryffindor_hist)), Gryffindor_hist, width=1, align='edge', color='red', alpha=0.5)
plt.bar(range(len(Hufflepuff_hist)), Hufflepuff_hist, width=1, align='edge', color='orange', alpha=0.5)
plt.xticks(x, ['ARI', 'AST', 'HER', 'DADA', 'DIV', 'MS',
                'AR', 'HM', 'TRA', 'POT', 'CMC', 'CHA', 'FLY'])
plt.title('Students activity')
plt.xlabel('Hogwarts courses')
plt.ylabel('Number of students')
plt.legend(loc='upper left')
plt.show()
```
